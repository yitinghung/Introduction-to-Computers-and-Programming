```python
import json
```

```python
def load_dictionary_from_json(filename):
    with open(filename) as f:
        return json.load(f)

def save_dictionary_from_json(data, filename):
    with open(filename, 'w') as f:
        json.dump(data, f)
```

```python
kamigo = load_dictionary_from_json("kamigo.json")
while(1):
    ans = input()
    if ans in kamigo:
        print("卡咪狗：", kamigo[ans])
        
    elif ans == "卡咪狗再見":
        print("掰掰～")
        break
        
    elif ans == "卡咪狗 你會說什麼":
        print("卡咪狗： 這裡教我說：")
        count = 0
        for i in kamigo.items():
            print(i[0], i[1], sep = '；')
            count += 1
            if count != len(kamigo):
                print("----------")
            
    elif ans.split("；")[0] == "卡咪狗學":
        kamigo[ans.split("；")[1]] = ans.split("；")[2]
        save_dictionary_from_json(kamigo, "kamigo.json")
        print("卡咪狗： 好哦～好哦～")
        
    elif ans.split("；")[0] == "卡咪狗忘記":
        kamigo.pop(ans.split("；")[1])
        save_dictionary_from_json(kamigo, "kamigo.json")
        print("卡咪狗： 好哦～好哦～")
        
    else:
        print("卡咪狗： 真的是拿你沒辦法呢...")
        print("卡咪狗： 你可以這樣教我說話")
        print("卡咪狗： 卡咪狗學；你好嗎；我很好")
    
```
